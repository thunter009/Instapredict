```python
import pandas as pd


path = '/Users/arianiherrera/Desktop/NYCDataScience/InstaPredict_Capstone/image_posts_df.csv'

insta_combined_df = pd.read_csv(path)
user_data = pd.read_csv('/Users/arianiherrera/Downloads/users.csv')
insta_combined_df.shape
```

```python
user_data.head()
```

```python
# features to drop from training set

insta_combined_df = insta_combined_df.loc[insta_combined_df['tag'] == 'nature']
insta_combined_df['user'] = insta_combined_df['user'].apply(lambda x: x[1:])
final_df = pd.merge(insta_combined_df, user_data, how='inner', left_on = 'user', right_on = 'user')
```

```python
final_df.shape
```

```python
# need to edit engineering file
from sklearn.preprocessing import LabelEncoder
from textblob import TextBlob
le = LabelEncoder()
final_df['filters'] = le.fit_transform(final_df['filters'])
```

```python
final_df['num_hashtags'] = final_df.hashtags.apply(lambda x: len(x))
final_df['hashtags'] = final_df.hashtags.apply(lambda x: ''.join(x))
final_df['hashtag_polarity'] = final_df.hashtags.apply(lambda x: TextBlob(x).sentiment.polarity)
final_df['hashtag_subjectivity'] = final_df.hashtags.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
feats_to_drop = ['Unnamed: 0_x', 'X1', 'caption', 'image', 'key', 'new_id', 'photo_id',
                  'scrape_time', 'index', 'image_key', 'post_time', 'time', 'likes','Unnamed: 0',
                 'bio', 'full_name', 'list_likes', 'top_N_dic', 'top_N_photos', 'hashtags', 'tag'
                 ]
from sklearn.cross_validation import train_test_split
X = final_df.drop(feats_to_drop, axis=1)
y = final_df[['likes']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.drop(['scrape_time_x','user', 'Unnamed: 0_y','scrape_time_y', 'Unnamed: 0.1'],axis=1,inplace=True)
X_test.drop(['scrape_time_x','user', 'Unnamed: 0_y', 'scrape_time_y', 'Unnamed: 0.1'],axis=1,inplace=True)
```

```python
print X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
```

```python
# create model and tune epochs and batch size parameters

def test_model():
    
    #define parameters
    input_num = 42
    hidden1_num = 100
    hidden2_num = 100
    hidden3_num = 100
    hidden4_num = 100
    hidden5_num = 100
    output_num = 1
    
    dropout_ratio = 0.5
    
    model = Sequential()
    model.add(Dense(output_dim=hidden1_num, input_dim=input_num, init='normal', activation='relu'))
    model.add(Dropout(dropout_ratio))
    model.add(Dense(output_dim=hidden2_num, input_dim=hidden1_num, init='normal', activation='relu'))
    model.add(Dropout(dropout_ratio))
    model.add(Dense(output_dim=hidden3_num, input_dim=hidden2_num, init='normal', activation='relu'))
    model.add(Dropout(dropout_ratio))
    model.add(Dense(output_dim=hidden4_num, input_dim=hidden3_num, init='normal', activation='relu'))
    model.add(Dropout(dropout_ratio))
    model.add(Dense(output_dim=hidden5_num, input_dim=hidden4_num, init='normal', activation='relu'))
    model.add(Dropout(dropout_ratio))
    model.add(Dense(1, init='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
    
```

```python
model = KerasRegressor(build_fn=test_model, nb_epoch = 10, batch_size = 60, verbose = 3)

# define thre grid search parameters
batch_size = [10, 20, 40, 60, 80, 100]
epochs = [10, 50, 100]
param_grid = dict(batch_size=batch_size, nb_epoch=epochs)

grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=3)

grid_result = grid.fit(X_train.as_matrix(), y_train.as_matrix())

print('Best: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
```
