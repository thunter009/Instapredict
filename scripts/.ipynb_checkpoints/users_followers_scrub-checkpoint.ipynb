```python
import pandas as pd
import numpy as np 
import ast
from scrub import scrub
```

```python
Mar17 = pd.read_csv('../data/Mar17_users_followers.csv')
Mar17
```

```python
Mar17['top_N_dic'] = [ast.literal_eval(x) for x in Mar17['top_N_photos']]
Mar17['list_likes'] = [x['likes'] for x in Mar17['top_N_dic']]
```

```python
Mar17['mean_likes'] = [np.mean(x) if len(x) > 0 else 0 for x in Mar17['list_likes']]
Mar17['median_likes'] = [np.median(x) if len(x) > 0 else 0 for x in Mar17['list_likes']]
Mar17['max_likes'] = [np.max(x) if len(x) > 0 else 0 for x in Mar17['list_likes']]
Mar17['min_likes'] = [np.min(x) if len(x) > 0 else 0 for x in Mar17['list_likes']]
```

```python
Mar17
```

```python
Mar17.to_csv('../data/Mar17_users_followers.csv')
```
